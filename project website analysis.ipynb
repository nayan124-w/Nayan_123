```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("website analysis.csv")
```

```python
data.columns
```

```python
data.columns = ['channel group', 'DateHour', 'Users', 'Sessions', 'Engaged Session', 'Average engagement time per second','Event Session per user','Event  per session','Engagement Rate','Event rate']
```

```python
data.columns
```

```python
data.head()
```

```python
data.columns = data.iloc[0]
```

```python
data.head()
```

```python
data = data.iloc[1:].reset_index(drop=True)
```

```python
data.head()
```

```python
data.info()
```

```python
data.rename(columns={'Session primary channel group (Default channel group)': 'channel group', 'Date + hour (YYYYMMDDHH)': 'Date + hour'}, inplace=True)
```

```python
data['Date + hour'] = pd.to_datetime(data['Date + hour'], format = "%Y%m%d%H")
```

```python
data.head()
```

```python
numeric_cols = data.columns.drop(["channel group", "Date + hour"])
data[numeric_cols] = data[numeric_cols].apply(pd.to_numeric, errors='coerce')
data["Hour"] = data["Date + hour"].dt.hour
```

```python
type(numeric_cols)
```

```python
data.head()
```

```python
data['date'] = data['Date + hour'].dt.date
```

```python
data.head()
```

```python
data.drop(['Date + hour'], axis=1,inplace= True)
```

```python
data.head()
```

```python
data.info()
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data.info()
```

# sessions and user over time

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,5))
data.groupby("date")[["Sessions","Users"]].sum().plot(ax=plt.gca())
plt.title("Sessions and users over time")
plt.xlabel("Date")
plt.ylabel("count")
plt.show()
```

# total users by channel

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Users', y='channel group', hue='channel group', data=data, palette='Set1')
plt.title('total users by channel')
plt.xlabel('Users')
plt.ylabel('channel group')
plt.show()
```

# Average engagement time by channel

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Average engagement time per session', y='channel group', hue='channel group', data=data, palette='Set1')
plt.title('Average engagement time by channel')
plt.xlabel('Average engagement time per session')
plt.ylabel('channel group')
plt.show()
```

# Engagement Rate Distribution by channel

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="channel group", y="Engagement rate", hue='channel group', data=data, palette='Set1')
plt.title(" Engagement Rate Distribution by Channel")
plt.xticks(rotation=45)
plt.show()
```

# Engaged vs non engaged sessions

```python
data.info()
```

```python
session_data = data.groupby("channel group")[["Sessions", "Engaged sessions"]].sum()
session_data["Non-Engaged"] = session_data["Sessions"] - session_data["Engaged sessions"]

# Reset index so that "channel group" wapas column ban jaye
session_data = session_data.reset_index()

# Melt karne ke liye ab column available hai
session_data_melted = session_data.melt(
    id_vars="channel group",
    value_vars=["Engaged sessions", "Non-Engaged"],
    var_name="Session Type",
    value_name="Count"
)

plt.figure(figsize=(8,5))
sns.barplot(data=session_data_melted, x="channel group", y="Count", hue="Session Type")
plt.title("Engaged vs Non-Engaged Sessions by Channel")
plt.xticks(rotation=45)
plt.show()
```

# traffic by hour and channel

```python
data.drop(['not_engaged_sessions','Non Engaged sessions','non_engaged_sessions'], axis = 1,inplace = True)
```

```python
data.head()
```

```python
heatmap_data = data.groupby(["Hour", "channel group"])["Sessions"].sum().unstack()
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=.5, annot=True, fmt='.0f')
plt.title(" Traffic by Hour and Channel")
plt.xlabel("Channel Group")
plt.ylabel("Hour of Day")
plt.show()
```

# engagement rate vs sessions over time

```python
# Group data
data_plot = data.groupby("date")[["Engagement rate", "Sessions"]].mean().reset_index()

# Create figure and axis
fig, ax1 = plt.subplots(figsize=(10,5))

# Engagement rate on left y-axis
ax1.plot(data_plot["date"], data_plot["Engagement rate"], color="green", label="Engagement rate")
ax1.set_xlabel("date")
ax1.set_ylabel("Engagement rate (%)", color="green")
ax1.tick_params(axis="y", labelcolor="green")

# Sessions on right y-axis
ax2 = ax1.twinx()
ax2.plot(data_plot["date"], data_plot["Sessions"], color="blue", label="Sessions")
ax2.set_ylabel("Sessions", color="blue")
ax2.tick_params(axis="y", labelcolor="blue")

# Title
plt.title(" Engagement Rate vs Sessions Over Time")

# Rotation for readability
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
